```python
import yfinance as yf
import pandas as pd

# Define the ticker symbols for Nifty sector-based indices
tickers = {
    'Nifty Auto': '^CNXAUTO',
    'Nifty Bank': '^CNXBANK',
    'Nifty Financial Services': '^CNXFIN',
    'Nifty FMCG': '^CNXFMCG',
    'Nifty IT': '^CNXIT',
    'Nifty Media': '^CNXMEDIA',
    'Nifty Metal': '^CNXMETAL',
    'Nifty Pharma': '^CNXPHARMA',
    'Nifty Realty': '^CNXREALTY',
    'Nifty 50': '^NSEI'
}

# Define the start and end date for the data
start_date = '2011-12-31'
end_date = '2021-12-31'

# Create a dictionary to hold the dataframes for each index
data_frames = {}

# Download the data for each ticker
for index_name, ticker in tickers.items():
    df = yf.download(ticker, start=start_date, end=end_date)
    data_frames[index_name] = df

# Save each dataframe to a separate CSV file
file_paths = []
for index_name, df in data_frames.items():
    file_name = f"{index_name.replace(' ', '_')}_10Y.csv"
    df.to_csv(file_name)
    file_paths.append(file_name)
    print(f"Saved {index_name} data to {file_name}")

file_paths
```

```python
import pandas as pd
# Path to the CSV file
file_path = '/content/india-gdp-growth-rate (1).csv'
gdp = pd.read_csv(file_path)
gdp['date'] = pd.to_datetime(gdp['date'])
start_date = '2011-12-31'
end_date = '2021-12-31'

# Filter the GDP data based on the date range
gdp = gdp[(gdp['date'] >= start_date) & (gdp['date'] <= end_date)]

gdp.set_index('date', inplace=True)
```

```python
# Path to the CSV file
file_path = '/content/india-inflation-rate-cpi.csv'
inflation = pd.read_csv(file_path)
inflation['date'] = pd.to_datetime(inflation['date'])
start_date = '2011-12-31'
end_date = '2021-12-31'

# Filter the inflation data based on the date range
inflation= inflation[(inflation['date'] >= start_date) & (inflation['date'] <= end_date)]

inflation.set_index('date', inplace=True)
print(inflation.head())
```

```python
# Path to the CSV file
file_path = '/content/india-unemployment-rate.csv'
unemployment = pd.read_csv(file_path)

unemployment['date'] = pd.to_datetime(unemployment['date'])
start_date = '2011-12-31'
end_date = '2021-12-31'

# Filter the unemployment data based on the date range
unemployment= unemployment[(unemployment['date'] >= start_date) & (unemployment['date'] <= end_date)]

unemployment.set_index('date', inplace=True)
print(unemployment.head())
```

```python
# Path to the CSV file
file_path = '/content/india-exports.csv'
exports = pd.read_csv(file_path)


exports['date'] = pd.to_datetime(exports['date'])

start_date = '2011-12-31'
end_date = '2021-12-31'
# Filter the exports data based on the date range
exports  = exports[(exports['date'] >= start_date) & (exports['date'] <= end_date)]

exports.set_index('date', inplace=True)
print(exports.head())
```
